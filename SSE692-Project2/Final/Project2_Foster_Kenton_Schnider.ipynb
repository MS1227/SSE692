## Distributed Version Control System (DVCS) <a class="anchor" id="dvcs"></a>
As technology and software engineering has evolved over time, development teams have grown, and projects have grown. Nowadays, development teams consist of two types of members – developers/engineers and testers. With this growth, there has been a growing need and emphasis for accurate documentation regarding software changes that an individual has made. Additionally, there has been an even bigger need, the ability to allow many different developers to work in parallel on one project without having to manually combine multiple copies of software. These two needs have been solved through version control systems, particularly Distributed Version Control Systems (DVCS).
DVCS can be thought of as a library, the software as books, and the user the reader in the following original analogy. Like a library, DVCS holds many pieces of software that can be checked out by users, read or edited for a period, and then checked back in to the DVCS. However, there are some differences, such as the “books” in this type of library often do not come back the same as they left, because they have been edited or rewritten. Another difference is that these edits and rewrites are documented when turned in, so the next reader will know what has changed. In DVCS, this log of changes is kept throughout the life cycle of the book, so that the testers of the development team know which parts of the software need to be changed. There are many DVCS systems a few popular ones being, GitHub, Team Foundation, and Bitbucket. All these systems are implementations of Git, which in and of itself is a DVCS.

### DVCS Structure <a class="anchor" id="dvcs_structure"></a>
DVCS has three main parts: the master repository, the local repository, and the working copy explained below.

*   **Shared Central Repository**
    *	The main directory where contributors can clone, branch, or pull software from
    *	Considered the “master” directory or branch of the directory
    *	Managed by a maintainer who accepts or rejects the changes from the contributors after they have reviewed the changes

*   **Local Repository**
    *	A per-user sandbox where they pull from the shared central repo, make changes, then push the changes back to the central repo for review and acceptance. 
    *	This also serves as a copy of the central repo as a backup

*   **Working Copy**
    *	File or files from the local repo that the user is actively using, modifying, or writing

### Repository Actions <a class="anchor" id="repo_actions"></a>
To interact with a repository, there are many actions that I have categorized into three different categories – States of a File in DVCS, Repository Management, and Requests.

*	**States of a File in a DVCS**
    *	Modified
        *	This is the state in which a user has made changes to a file, but the changes have not been committed
        *	Occurs is in the Working Directory shown in Figure 2.
    *	Staged
        *	This is the state in which the modified file has been marked and will be committed
        *	Occurs is in the Staging Area shown in Figure 2.
    *	Committed
        *	This is the state in which the file is stored in a user’s local repository or the master repository

Below is an image that shows the states and transitions a file can experience in version control.

>![states_of_file_in_repo.png](https://git-scm.com/figures/18333fig0106-tn.png)

>[Link to Image](https://git-scm.com/book/en/v1/Getting-Started-Git-Basics)


*	**Repository Management**
    *	Initialize
        *	Initialization of a repository can be done as the master or as a sub-repository within an existing repository
    *	Cloning
        *	Cloning a repository is the act of making a copy of the repository (project) with the intention of being a contributor to the project
        *	Cloning can occur on the master or any sub-repository
    *	Branch
        *	Branching is similar cloning, but this branch is where new changes occur to smaller branch of the project and not the whole project
    *	Merge
        *	Merging is the opposite of branching, so it combines two branches in order into one branch merging their changes

*	**Requests**
    *	Push
        *	Pushing of files is the movement of committed files to their local repository and the shared central repository
    *	Pull
        *	Pulling files is the opposite of the pushing, so the user pulls files from the shared central repository to edit

### Step 4 - Prepare the data for Machine Learning algorithms.

The final stage of cleaning consists of two steps: removing unwanted features and creating train/test datasets. First, create two new pandas dataframes with only the three deatures that will be used for regression - **RM**, **LSTAT**, and **MEDV**. The **X** axis (dataframe) will be **RM** and **LSTAT** and the **Y** axis (dataframe) will be the target feature **MEDV**.  The code that demonstrates this can be seen below.

````python
x_df = data
for feature in data:
  if feature == 'LSTAT' or feature == 'RM':
    pass
  else:
    x_df = x_df.drop(feature, axis=1)
    
y_df = data['MEDV']
````

Now that the **X** and **Y** dataframes have been created, its time to divide the dataframes into training data and testing data. ### Linear Regression
As previously mentioned, the first model to be used for deciding the cost of each house is the SGD Regressor. In order to approach the SGD Regressor with the right midset, a Linear Regression model will be trained and used to predict the house prices in the Boston Housing dataset. 

The Linear Regression will be the starting point for predicting the cost of houses in the Boston Housing dataset. In order to implement an Linear Regression model, scikit-learn will be used. The steps for creating, training, and testing are as follows:

1.   Create a **Linear Regression object** using scikit learn.
2.   **Fit the X-axis and Y-axis training data** with the Linear Regression model.
3.   Make predictions from the **X-axis of the test dataset**.
4.   Finally, the **accuracy of the Linear Regression model** will be recorded.

The code that demonstrates this can be seen below.

````python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()

lin_reg.fit(x_train, y_train)

y_pred = lin_reg.predict(x_test)

lin_reg_accuracy = round(lin_reg.score(x_train, y_train) * 100, 2)

print(lin_reg_accuracy, " %")
````

The result of this Linear Regression model is ~63%, which is decent but not great. Lets see how the model is preforming using cross validation.

#### Cross Validation Score
Cross validation is a technique that checks to see if the model is actually performing as it seems. In order to cross validate the Linear Regression Model, a **K-Fold Cross Validation** will be used. This specific technique splits the training data into *K* subsets, trains the model *K-1* times, and then evaluates it using the last subset. For this Linear Regression Model, the number of folds is **K = 10**. Scikit Learn provides a nice function called **cross_val_score**, which will be used. The code that demonstrates this can be seen below.

````python
from sklearn.model_selection import cross_val_score

lin_reg = LinearRegression()

lin_reg.fit(x_train, y_train)

lin_reg_score = cross_val_score(lin_reg, x_train, y_train, cv=10)

print("Mean Accuracy: {} %\n\nStd. Dev. Accuracy: {} %".format(round(lin_reg_score.mean() * 100, 2), round(lin_reg_score.std() * 100, 2)))
````

Based off the cross validation the average accuracy score was ~61% with a standard deviation of ~10%. Based off this metric, the algorithm is performing slightly less than what was originally believed to be the case. Another metric that is import to evaluate is the R<sup>2</sup> score.

#### R<sup>2</sup> Score
The R<sup>2</sup> score or the score of the coefficient of determination (COD) is a good test of accuracy for a model. In order to assess the COD, the model must make some predictions. This is accomplished using **cross_val_predict** from sckit learn using **cv=10** meaning 10 folds of the data will be used in the cross validation. Once predictions have been made, **r2_score** will be used from the scikit learn metric library to score the accuracy of the model. First, lets check the R<sup>2</sup> Score for the training data. The code that demonstrates this can be seen below.

````python
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import r2_score

lin_reg = LinearRegression()

lin_reg.fit(x_train, y_train)

lin_reg_pred = cross_val_predict(lin_reg, x_train, y_train, cv=10)

accuracy = r2_score(y_train, lin_reg_pred)

print("Training Coefficient of Determination: {} %\n".format(round(accuracy * 100, 2)))
````

Now lets check the R<sup>2</sup> Score for the test data. The code that demonstrates this can be seen below.

````python
lin_reg_pred = cross_val_predict(lin_reg, x_test, y_test, cv=10)

accuracy = r2_score(y_test, lin_reg_pred)

print("Test Coefficient of Determination: {} %\n".format(round(accuracy * 100, 2)))
````

The training data produced a COD of ~62%, which is approximately equivalent to the average accuracy achieved, so no surprises here. However, when ran on the test data, a significant increase has occurred to the COD! The COD increased from ~62% to ~71%, which is a good sign. Next, lets test the Mean Squared Error and the Root Mean Squared Error.

#### Mean Squared Error (MSE) and Root MSE (RMSE)
Another import metric of validating the accuracy of a model is the MSE and RMSE. MSE is average error between the actual value and estimation. The RMSE is simply the sqaure root of the MSE. Due to the importance of this metric, scikit learn provides a function for calculating the MSE in the metrics library and then the RMSE will be calculated by taking the root of this function. First, lets check the MSE and RMSE for the training data. The code that demonstrates this can be seen below.

````python
from sklearn.metrics import mean_squared_error

y_pred = lin_reg.predict(x_train)

mse = mean_squared_error(y_train, y_pred)

rmse = np.sqrt(mse)

print("Training MSE: {} %\n\nTraining RMSE: {} %".format(round(mse, 2), round(rmse, 2)))
````

Now lets check the MSE and RMSE for the test data. The code that demonstrates this can be seen below.

````python
y_pred = lin_reg.predict(x_test)

lin_reg_test_mse = mean_squared_error(y_test, y_pred)

lin_reg_test_rmse = np.sqrt(lin_reg_test_mse)

print("Test Data MSE: {} %\n\nTest Data RMSE: {} %".format(round(lin_reg_test_mse, 2), round(lin_reg_test_rmse, 2)))
````

The training data produced an MSE of ~32% and an RMSE of ~5.6%, when the test data was test the MSE decreased to ~26.5% and an RMSE of ~5.1%. Based of these reults, the model is behaving appropriately and has learned! Now that we have assessed a Linear Regression model, lets see how an SGD Regressor compares, so that an informed final decision on which model to choose can be made.

### SGD Regressor
As previously mentioned, the first model to be used for deciding the cost of each house is the SGD Regressor. SGD stands for Stochastic Gradient Descent, so this model is a Stochastic Gradient Descent Regressor. An SGD Regressor is an improved Linear Regression model. This model implements an algorithm that seeks to minimize loss through use of gradient descent. Minimization of loss is achieved by updating the model with the estimated gradient of the for the current sample. The idea of SGD Regression is similar to a feed-forward loop or feedback loop, depending on how one looks at it. As mentioned previously, the SGD Regressor seeks to improve upon the Linear Regression model, so lets put that to the test.

The SGD Regressor will be the starting point for predicting the cost of houses in the Boston Housing dataset. In order to implement an SGD Regressor, scikit-learn will be used. The steps for creating, training, and testing are as follows:

1.   Create a **SGD Regressor object** using scikit learn.
2.   **Fit the X-axis and Y-axis training data** with the SGD Regressor model.
3.   Make predictions from the **X-axis of the test dataset**.
4.   Finally, the **accuracy of the SGD Regressor model** will be recorded.

The code that demonstrates this can be seen below.

````python
from sklearn.linear_model import SGDRegressor

sgd_reg = SGDRegressor(eta0=0.1, max_iter=1000, tol=1e-3)

sgd_reg.fit(x_train, y_train)

sgd_reg_accuracy = round(sgd_reg.score(x_train, y_train) * 100, 2)

print(sgd_reg_accuracy, " %")
````

With an accuracy score of "-6.1788153294070294e+22  %", this raises a lot of red flags. Lets check the MSE and RMSE to verify that something is in fact wrong. The code that demonstrates this can be seen below.

````python
from sklearn.metrics import mean_squared_error

y_pred = sgd_reg.predict(x_train)

mse = mean_squared_error(y_train, y_pred)

rmse = np.sqrt(mse)

print("Training MSE: {} %\n\nTraining RMSE: {} %".format(round(mse, 2), round(rmse, 2)))
````

The suspicions are confirmed. We appear to be moving in the wrong direction - by a lot! At this point, a GridSearchCV should be performed on the hyperparameters as to reduce and/or eliminate useless hyperparameters.

### Step 6 - Fine-tune your model - Part 1

#### Tuning the Hyperparameters - SGD Regressor
Hyperparameters are parameters within a machine learning model that can vary and produce completely different results within the model. Ideally, all of the hyperparameters would be set exactly to the right value before hand, so that the model is trained optimally. In order to determine the best value to assign to these hyperparameters, a GridSearchCV will be applied to the following hyperparameters:
*   alpha: Constant that multiplies the regularization term. Defaults to 0.0001 Also used to compute learning_rate when set to ‘optimal’.
*   loss: The loss function to be used. The possible values are ‘squared_loss’, ‘huber’, ‘epsilon_insensitive’, or ‘squared_epsilon_insensitive’.
*   penalty: The penalty (aka regularization term) to be used.
*   learning_rate: The learning rate schedule, which can be 'constant', ‘optimal’, ‘invscaling’, or ‘adaptive’. These are each different equations that manipulate the rate at which the model can learn.

An arbitrary set of values has been chosen for each of the hyperparameters and was compiled into a dictionart called ***param_grid***. GridSearchCV will cross validate the hyperparameters and return the optimal values for the model.

The code that demonstrates this can be seen below.

[Reference for SGD Regressor Hyperparameter Definitions](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.SGDRegressor.html#sklearn.linear_model.SGDRegressor)

````python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'alpha': 10.0 ** -np.arange(1, 7),
    'loss': ['squared_loss', 'huber', 'epsilon_insensitive'],
    'penalty': ['l2', 'l1', 'elasticnet'],
    'learning_rate': ['constant', 'optimal', 'invscaling'],
}

sgd_reg = SGDRegressor(max_iter=1000, tol=1e-3)

sgd_reg_clf = GridSearchCV(estimator=sgd_reg, param_grid=param_grid, n_jobs=-1, cv=10, iid=True)

sgd_reg_clf.fit(x_train, y_train)

sgd_reg_clf.best_params_
````

Based off the GridSearchCV, the hyperparameters for the SGD Regressor should be updated as follows:
*   alpha = 1e-05
*   loss = 'epsilon_insensitive'
*   penalty = 'elasticnet'
*   learning_rate = 'invscaling'

The code that demonstrates this can be seen below.

````python
sgd_reg = SGDRegressor(**sgd_reg_clf.best_params_, max_iter=1000, tol=1e-3)

sgd_reg.fit(x_train, y_train)

sgd_reg_accuracy = round(sgd_reg.score(x_train, y_train) * 100, 2)

print(sgd_reg_accuracy, " %")
````

With an accuracy score of ~61%, lets look at the other validation tests that haven't been ran yet such as Cross Validation, R^2 Score, MSE, and RMSE.

#### Cross Validation Score
As previously mentioned, a **K-Fold Cross Validation** will be used to validate the SGD Regressor. This specific technique splits the training data into *K* subsets, trains the model *K-1* times, and then evaluates it using the last subset. For this SGD Regressor, the number of folds is **K = 10**. Scikit Learn provides a nice function called **cross_val_score**, which will be used. The code that demonstrates this can be seen below.

````python
from sklearn.model_selection import cross_val_score

sgd_reg = SGDRegressor(**sgd_reg_clf.best_params_, max_iter=1000, tol=1e-3)

sgd_reg.fit(x_train, y_train)

sgd_reg_score = cross_val_score(sgd_reg, x_train, y_train, cv=10)

print("Mean Accuracy: {} %\n\nStd. Dev. Accuracy: {} %".format(round(sgd_reg_score.mean() * 100, 2), round(sgd_reg_score.std() * 100, 2)))
````

#### R<sup>2</sup> Score
As mentioned previously, the R<sup>2</sup> score or the score of the coefficient of determination (COD) is a good test of accuracy for a model. First, lets check the R<sup>2</sup> Score for the training data using the SGD Regressor. The code that demonstrates this can be seen below.

````python
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import r2_score

sgd_reg = SGDRegressor(**sgd_reg_clf.best_params_, max_iter=1000, tol=1e-3)

sgd_reg.fit(x_train, y_train)

sgd_reg_pred = cross_val_predict(sgd_reg, x_train, y_train, cv=10)

accuracy = r2_score(y_train, sgd_reg_pred)

print("Training Coefficient of Determination: {} %\n".format(round(accuracy * 100, 2)))
````

Now lets check the R<sup>2</sup> Score for the test data using the SGD Regressor. The code that demonstrates this can be seen below.

````python
sgd_reg_pred = cross_val_predict(sgd_reg, x_test, y_test, cv=10)

accuracy = r2_score(y_test, sgd_reg_pred)

print("Test Coefficient of Determination: {} %\n".format(round(accuracy * 100, 2)))
````

The training data produced a COD of ~61%, which is slightly higher than average accuracy of ~59%. When ran on the test data, an increase from ~62% to ~71% has occurred, significant but not as impressive as the Linear Regression Model. #### Mean Squared Error (MSE) and Root MSE (RMSE)

### Step 6 - Fine-tune your model - Part 2

#### Tuning the Hyperparameters - Linear Regression
Hyperparameters are parameters within a machine learning model that can vary and produce completely different results within the model. Ideally, all of the hyperparameters would be set exactly to the right value before hand, so that the model is trained optimally. In order to determine the best value to assign to these hyperparameters, a GridSearchCV will be applied to the following hyperparameters:
*   fit_intercept: whether to calculate the intercept for this model. If set to False, no intercept will be used in calculations (e.g. data is expected to be already centered).
*   normalize: This parameter is ignored when fit_intercept is set to False. If True, the regressors X will be normalized before regression by subtracting the mean and dividing by the l2-norm. If you wish to standardize, please use sklearn.preprocessing.StandardScaler before calling fit on an estimator with normalize=False.
*   copy_X: If True, X will be copied; else, it may be overwritten.

An arbitrary set of values has been chosen for each of the hyperparameters and was compiled into a dictionart called ***param_grid***. GridSearchCV will cross validate the hyperparameters and return the optimal values for the model.

The code that demonstrates this can be seen below.

[Reference for Linear Regression Hyperparameter Definitions](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html)

````python
param_grid = {
    'fit_intercept':[True,False],
    'normalize':[True,False],
    'copy_X':[True, False],
}

lin_reg_2 = LinearRegression()

lin_reg_clf_2 = GridSearchCV(estimator=lin_reg_2, 
                           param_grid=param_grid, 
                           n_jobs=-1, 
                           cv=10, 
                           iid=True)

lin_reg_clf_2.fit(x_train, y_train)

lin_reg_clf_2.best_params_
````

Based off the GridSearchCV, the hyperparameters for the Linear Regression Model should be updated as follows:
*   fit_intercept = False
*   normalize = True
*   copy_X = True

As such, the hyperparameters and all of the validations have been ran again.

The code that demonstrates this can be seen below.

````python
lin_reg_2 = LinearRegression(**lin_reg_clf_2.best_params_)

lin_reg_2.fit(x_train, y_train)

y_pred = lin_reg_2.predict(x_test)

#Cross Validation Score
lin_reg_score = cross_val_score(lin_reg_2, x_train, y_train, cv=10)

print("Mean Accuracy: {} %\n\nStd. Dev. Accuracy: {} %\n".format(round(lin_reg_score.mean() * 100, 2), round(lin_reg_score.std() * 100, 2)))

#R^2 Score
lin_reg_pred = cross_val_predict(lin_reg_2, x_test, y_test, cv=10)

lin_reg_accuracy = r2_score(y_test, lin_reg_pred)

print("Test Coefficient of Determination: {} %\n".format(round(lin_reg_accuracy * 100, 2)))

lin_reg_accuracy = lin_reg_accuracy * 100

#MSE and RMSE
lin_reg_test_mse = mean_squared_error(y_test, y_pred)

lin_reg_test_rmse = np.sqrt(lin_reg_test_mse)

print("Test Data MSE: {} %\n\nTest Data RMSE: {} %".format(round(lin_reg_test_mse, 2), round(lin_reg_test_rmse, 2)))
````
